```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sc_bc_act = np.loadtxt('sc_broadcast_activity.txt', dtype=np.uint64)
sc_dp_act = np.loadtxt('sc_dispatch_activity.txt', dtype=np.uint64)
sc_di_act = np.loadtxt('sc_discard_activity.txt', dtype=np.uint64)
sc_lat = np.loadtxt('sc_latency.txt', dtype=np.uint64)
sc_cong = np.loadtxt('sc_congestion.txt', dtype=np.uint64)
```

```python
fig, ax = plt.subplots(dpi=150)
ax.hist(sc_bc_act, bins=400, density=True, alpha=0.5, label='broadcast activity')
ax.hist(sc_dp_act, bins=400, density=True, alpha=0.5, label='dispatch activity')
ax.hist(sc_di_act, bins=400, density=True, alpha=0.5, label='discard activity')
ax.legend(loc='upper right')
ax.set_xlabel('activities density histogram')
fig.savefig('activity.png', dpi=600)
```

```python
fig, ax = plt.subplots(dpi=150)
ax.scatter(sc_lat[:, 0], sc_lat[:, 1], s=0.1)
ax.set_xlabel('time')
ax.set_ylabel('latency')
fig.savefig('latency.png', dpi=600)
```

```python
fig, ax = plt.subplots(dpi=150)
ax.scatter(sc_cong[:, 0], sc_cong[:, 1], s=0.1)
ax.set_xlabel('time')
ax.set_ylabel('congestion')
fig.savefig('congestion.png', dpi=600)
```

```python
# plot orchestration distribution (half circle dist: (sin(x * PI - PI / 2) + 1) / 2) from 0 to 1
fig, ax = plt.subplots(dpi=150)
x = np.linspace(0, 1, 10000)
y = (np.sin(x * np.pi - np.pi / 2) + 1) / 2
ax.plot(x, y)
ax.set_xlabel('orchestration distribution')
fig.savefig('orchestration.png', dpi=600)
```

```python
# plot orchestration distribution's derivative
fig, ax = plt.subplots(dpi=150)
x = np.linspace(0, 1, 10000)
y = np.cos(x * np.pi - np.pi / 2) * np.pi / 2
ax.plot(x, y)
ax.set_xlabel('orchestration distribution derivative')
fig.savefig('orchestration_derivative.png', dpi=600)
```
